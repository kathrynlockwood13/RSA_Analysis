```python
#1.encoding 1, rest1 day 1
#corr plot generation for loc, ffa, and ppa 
import pandas as pd
import numpy as np
import os

# Define a dictionary to map regions to their corresponding trial input files
region_trial_files = {
    'ffa': 'fus',
    'ppa': 'php',
    'loc': 'loc'  # Assuming 'loc' region uses the same trial input file for all subjects
}

# Loop through subjects '02' to '30'
for subject_num in range(2, 31):
    participant_str = f'{subject_num:02}'  # Format subject number with leading zero if needed

    for region in region_trial_files.keys():
        # Read the CSV files into DataFrames
        trial_input_file = f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/LSS_voxXtrial_{region_trial_files[region]}_sub-{participant_str}_run-01.csv'
        tr_input_file = f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/LSS_voxXTR_{region}_sub-{participant_str}_run-01_day1.csv'
        
        df1 = pd.read_csv(trial_input_file)
        df2 = pd.read_csv(tr_input_file)

        # Drop the first column from df2
        df2 = df2.iloc[:, 1:]
        # Convert to numeric, ignoring errors
        df1 = df1.apply(pd.to_numeric, errors='coerce')
        df2 = df2.apply(pd.to_numeric, errors='coerce')

        # Concatenate the two dataframes along the columns (TRs first, then Trials)
        df = pd.concat([df2, df1], axis=1)

        # Compute the correlation matrix
        correlation_matrix = df.corr()

        # Determine the number of columns in each dataframe
        num_trial_columns = df1.shape[1]
        num_tr_columns = df2.shape[1]

        # Extract the portion of the correlation matrix corresponding to TR and Trial
        correlation_tr_trial = correlation_matrix.iloc[:num_tr_columns, num_tr_columns:]

        # Remove the 'Unnamed: 0' column
        correlation_tr_trial = correlation_tr_trial.drop(columns=['Unnamed: 0'])

        # Set row names to 'TR_01' through 'TR_210'
        correlation_tr_trial.index = [f'TR_{i:02}' for i in range(1, num_tr_columns + 1)]

        # Save correlation matrix as a CSV file with row names and column names
        output_folder = f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/{region}_corr'
        output_filename = os.path.join(output_folder, f'corr_matrix_{region}_TR_Run-01_day1_by_Trial-Run-01.csv')
        correlation_tr_trial.to_csv(output_filename, index=True, header=True)

    print(f"Correlation matrices saved for Subject {participant_str}")
```

```python
import pandas as pd
import numpy as np
import os

# Fixed subject and region
participant_str = '02'
region = 'ffa'

# Read just the header (first row) of the CSV to get the column names
column_names_file = f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/ffa_corr/corr_matrix_ffa_TR_Run-01_day1_by_Trial-Run-01.csv'
column_names = pd.read_csv(column_names_file, nrows=0).columns.tolist()

# Read the CSV files into DataFrames
df2 = pd.read_csv(f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/voxBYtr_{region}_sub-{participant_str}_run-02_day2rest.csv')
df1 = pd.read_csv(f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/voxBYtrial_{region}_sub-{participant_str}_run-01_day2enc.csv')

# Drop the first column from df2
df2 = df2.iloc[:, 1:]

# Convert to numeric, ignoring errors
df1 = df1.apply(pd.to_numeric, errors='coerce')
df2 = df2.apply(pd.to_numeric, errors='coerce')

# Concatenate the two dataframes along the columns (TRs first, then Trials)
df = pd.concat([df2, df1], axis=1)

# Compute the correlation matrix
correlation_matrix = df.corr()

# Determine the number of columns in each dataframe
num_trial_columns = df1.shape[1]
num_tr_columns = df2.shape[1]

# Extract the portion of the correlation matrix corresponding to TR and Trial
correlation_tr_trial = correlation_matrix.iloc[:num_tr_columns, num_trial_columns:]

# Apply the previously fetched column names
if len(correlation_tr_trial.columns) == len(column_names):
    correlation_tr_trial.columns = column_names

# Remove the 'Unnamed: 0' column if it exists
if 'Unnamed: 0' in correlation_tr_trial.columns:
    correlation_tr_trial = correlation_tr_trial.drop(columns=['Unnamed: 0'])

# Set row names to 'TR_01' through 'TR_210' (or whatever the number of TR columns is)
correlation_tr_trial.index = [f'TR_{i:02}' for i in range(1, num_tr_columns + 1)]

# Create output directory if it does not exist
output_folder = f'/data/Kathryn/Projects/Priority/derivatives/rsa_dfs/{region}/sub-{participant_str}/{region}_corr'
os.makedirs(output_folder, exist_ok=True)

# Save correlation matrix as a CSV file with row names and column names
output_filename = os.path.join(output_folder, f'corr_matrix_{region}_TR_Run-02_day2_by_Trial-Run-01_test.csv')
correlation_tr_trial.to_csv(output_filename, index=True, header=True)

print(f"Correlation matrix saved for Subject {participant_str} in Region {region}")
```
